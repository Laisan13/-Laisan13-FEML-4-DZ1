## Логистическая регрессия

```python
import numpy as np
import pandas as pd 
import scipy.stats 
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
data = pd.read_csv('adult_csv.csv')
data
```

```python
data.info()
```

```python
data = data[pd.isnull(data['native-country']) == 0 ] # исключаем нулевые поз
data.info()
```

```python
selectedColumns = data [['education-num', 'occupation','sex']]
#выбираем колонки образование,род занятий, пол
```

```python
X = pd.get_dummies (selectedColumns)
X
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit( data['class'])#переведем значения столбца class в числа, оставив один столбец
y = pd.Series (data = le.transform(data['class'])) 
y
```

```python
from sklearn.pipeline import make_pipeline 
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
model = make_pipeline(StandardScaler(), LogisticRegression(max_iter=1000))
# используем пайплайны для удобства, приводим в единую размерность 
# берем в качестве модели логистическую регресиию
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model.fit(X_train, y_train)
predictions = model.predict(X_test)
```

```python
predictions[:5]
```

```python
model.predict(X_test)
```

```python
model.predict_proba(X_test)
```

```python
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

## Метод опорных векторов. SVM.

```python
from sklearn.svm import SVC
clf = make_pipeline(StandardScaler(), SVC(gamma='auto')) 
clf.fit(X_train, y_train)
```

```python
clf.score(X_train, y_train) 
```

```python
clf.score(X_test, y_test)
```
